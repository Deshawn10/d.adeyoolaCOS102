```python
# Project III - Student Data Table
# Girls Data
girls_names = ["Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola", "Aisha", "Latifa"]
girls_age = [17, 16, 17, 18, 16, 18, 17, 20, 19, 17]
girls_height = [5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5]
girls_scores = [80, 85, 70, 60, 76, 66, 87, 95, 50, 49]

# Boys Data
boys_names = ["Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley"]
boys_age = [19, 16, 18, 17, 20, 19, 16, 18, 17, 19]
boys_height = [5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7]
boys_scores = [74, 87, 75, 68, 66, 78, 87, 98, 54, 60]

def print_table(title, names, ages, heights, scores):
    print("\n" + "=" * 60)
    print(f"{title:^60}")
    print("=" * 60)
    print(f"{'Name':<15} {'Age':<8} {'Height':<10} {'Score':<10}")
    print("-" * 60)
    
    for i in range(len(names)):
        print(f"{names[i]:<15} {ages[i]:<8} {heights[i]:<10} {scores[i]:<10}")
    
    print("-" * 60)
    
    # Calculate and display statistics
    avg_age = sum(ages) / len(ages)
    avg_height = sum(heights) / len(heights)
    avg_score = sum(scores) / len(scores)
    max_score = max(scores)
    min_score = min(scores)
    
    print(f"\nStatistics for {title}:")
    print(f"Average Age: {avg_age:.2f}")
    print(f"Average Height: {avg_height:.2f}")
    print(f"Average Score: {avg_score:.2f}")
    print(f"Highest Score: {max_score}")
    print(f"Lowest Score: {min_score}")

# Display tables
print_table("GIRLS DATA", girls_names, girls_age, girls_height, girls_scores)
print_table("BOYS DATA", boys_names, boys_age, boys_height, boys_scores)
```
