```python
# Project II - Find Roots of Equations
import math

def quadratic_roots(a, b, c):
    """Find roots of quadratic equation ax^2 + bx + c = 0"""
    if a == 0:
        return "Not a quadratic equation"
    
    discriminant = b**2 - 4*a*c
    
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return f"Two real roots: {root1:.2f} and {root2:.2f}"
    elif discriminant == 0:
        root = -b / (2*a)
        return f"One real root (double root): {root:.2f}"
    else:
        real_part = -b / (2*a)
        imag_part = math.sqrt(-discriminant) / (2*a)
        return f"Two complex roots: {real_part:.2f} + {imag_part:.2f}i and {real_part:.2f} - {imag_part:.2f}i"

def cubic_roots(a, b, c, d):
    """Find roots of cubic equation ax^3 + bx^2 + cx + d = 0"""
    if a == 0:
        return "Not a cubic equation"
    
    # Using numpy if available, otherwise use approximation
    try:
        import numpy as np
        coeff = [a, b, c, d]
        roots = np.roots(coeff)
        return f"Cubic roots: {roots}"
    except ImportError:
        return "Cubic equation solver requires numpy. Install with: pip install numpy"

def quartic_roots(a, b, c, d, e):
    """Find roots of quartic equation ax^4 + bx^3 + cx^2 + dx + e = 0"""
    if a == 0:
        return "Not a quartic equation"
    
    try:
        import numpy as np
        coeff = [a, b, c, d, e]
        roots = np.roots(coeff)
        return f"Quartic roots: {roots}"
    except ImportError:
        return "Quartic equation solver requires numpy. Install with: pip install numpy"

def main():
    print("=" * 50)
    print("ROOT FINDER FOR POLYNOMIAL EQUATIONS")
    print("=" * 50)
    print("\nChoose an operation:")
    print("1. Quadratic Equation (ax^2 + bx + c = 0)")
    print("2. Cubic Equation (ax^3 + bx^2 + cx + d = 0)")
    print("3. Quartic Equation (ax^4 + bx^3 + cx^2 + dx + e = 0)")
    
    choice = int(input("\nEnter your choice (1/2/3): "))
    
    if choice == 1:
        print("\n--- Quadratic Equation ---")
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        result = quadratic_roots(a, b, c)
        print(result)
    
    elif choice == 2:
        print("\n--- Cubic Equation ---")
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        d = float(input("Enter coefficient d: "))
        result = cubic_roots(a, b, c, d)
        print(result)
    
    elif choice == 3:
        print("\n--- Quartic Equation ---")
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        d = float(input("Enter coefficient d: "))
        e = float(input("Enter coefficient e: "))
        result = quartic_roots(a, b, c, d, e)
        print(result)
    
    else:
        print("Invalid choice! Please choose 1, 2, or 3.")

if __name__ == "__main__":
    main()
```
