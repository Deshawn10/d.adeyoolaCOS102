```python
# Project IV - Annual Tax Revenue Calculator
def calculate_atr(experience, age):
    """Calculate Annual Tax Revenue based on experience and age"""
    if experience > 25 and age >= 55:
        return 5600000
    elif experience > 20 and age >= 45:
        return 4480000
    elif experience > 10 and age >= 35:
        return 1500000
    else:
        return 550000

def main():
    print("=" * 60)
    print("IZIFIN TECHNOLOGY - ANNUAL TAX REVENUE CALCULATOR")
    print("=" * 60)
    
    num_employees = int(input("\nEnter the number of staff to process: "))
    
    print("\n" + "-" * 60)
    print(f"{'Staff':<10} {'Experience':<12} {'Age':<8} {'ATR':<15}")
    print("-" * 60)
    
    total_atr = 0
    
    for i in range(num_employees):
        print(f"\nStaff {i+1}:")
        experience = float(input("  Years of experience: "))
        age = float(input("  Age: "))
        
        atr = calculate_atr(experience, age)
        total_atr += atr
        
        print(f"  Staff {i+1} ATR: N{atr:,}")
    
    print("\n" + "=" * 60)
    print(f"Total Annual Tax Revenue for all staff: N{total_atr:,}")
    print(f"Average Annual Tax Revenue per staff: N{total_atr/num_employees:,.2f}")
    print("=" * 60)

if __name__ == "__main__":
    main()
```
