```python
# cell_2.py - Function that changes a list

def changeme(mylist):
    """This changes a passed List"""
    mylist.append([1, 2, 3, 4])
    print("Values inside the function: ", mylist)
    return

# Create a list and call the function
mylist = [10, 20, 30]
changeme(mylist)
print("Values outside the function: ", mylist)
```
