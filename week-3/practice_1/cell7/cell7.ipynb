```python
# cell_7.py - Function that can take multiple arguments

def printinfo(arg1, *vartuple):
    """This is test"""
    print("Output is: ")
    print(arg1)
    for var in vartuple:
        print(var)
    return

# Call with one argument
printinfo(10)

# Call with three arguments
printinfo(70, 60, 50)
```
