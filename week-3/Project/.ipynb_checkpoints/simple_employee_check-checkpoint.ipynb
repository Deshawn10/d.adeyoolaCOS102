```python
# simple_employee_check.py - Console version of employee checker

import csv

def load_employees():
    """Load employees from CSV file"""
    employees = {}
    
    try:
        with open('GIG-logistics.csv', 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                name = row['Name']
                department = row['Department']
                role = row['Role']
                
                employees[name] = {
                    'department': department,
                    'role': role
                }
        return employees
    except FileNotFoundError:
        print("CSV file not found. Using sample data...")
        # Sample data
        return {
            "Chukwuemeka Okafor": {"department": "Operations", "role": "Operations Manager"},
            "Adaeze Nwosu": {"department": "Operations", "role": "Dispatch Officer"},
            "Amina Usman": {"department": "Human Resources", "role": "HR Manager"},
            "Rotimi Adeyemi": {"department": "Finance", "role": "Accountant"}
        }

def main():
    print("=" * 50)
    print("GIG LOGISTICS - EMPLOYEE VERIFICATION SYSTEM")
    print("=" * 50)
    
    # Load employee data
    employees = load_employees()
    
    while True:
        print("\n" + "-" * 50)
        print("1. Check if someone is an employee")
        print("2. List all employees")
        print("3. Exit")
        print("-" * 50)
        
        choice = input("Enter your choice (1-3): ")
        
        if choice == "1":
            name = input("\nEnter employee name: ")
            
            if name in employees:
                emp = employees[name]
                print(f"\n✓ {name} is an employee!")
                print(f"  Department: {emp['department']}")
                print(f"  Role: {emp['role']}")
            else:
                print(f"\n✗ {name} is NOT an employee of GIG Logistics.")
        
        elif choice == "2":
            print("\n--- All Employees ---")
            for name, details in employees.items():
                print(f"• {name} - {details['department']}")
        
        elif choice == "3":
            print("\nThank you for using GIG Logistics Verification System!")
            break
        
        else:
            print("\nInvalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
