```python
# employee_verification.py - GIG Logistics Employee Checker
# This program checks if a person is an employee and shows department members

import tkinter as tk
from tkinter import messagebox
import csv

# Employee data from the CSV file
# We'll store employees in a dictionary for easy lookup
employees = {}

# Department members list
department_members = {
    "Operations": [],
    "Finance": [],
    "Human Resources": [],
    "Technology": [],
    "Customer Service": [],
    "Sales": [],
    "Marketing": []
}

def load_employees_from_csv():
    """Load employee data from the CSV file"""
    try:
        with open('GIG-logistics.csv', 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                name = row['Name']
                department = row['Department']
                role = row['Role']
                email = row['Email']
                location = row['Location']
                
                # Store employee info
                employees[name] = {
                    'department': department,
                    'role': role,
                    'email': email,
                    'location': location
                }
                
                # Add to department members list
                if department in department_members:
                    department_members[department].append(name)
        
        return True
    except FileNotFoundError:
        print("CSV file not found. Using sample data.")
        # Sample data if CSV not found
        add_sample_data()
        return False
    except Exception as e:
        print(f"Error loading CSV: {e}")
        add_sample_data()
        return False

def add_sample_data():
    """Add sample data if CSV file is not available"""
    sample_employees = [
        ("Chukwuemeka Okafor", "Operations", "Operations Manager", "c.okafor@giglogistics.com", "Lagos"),
        ("Adaeze Nwosu", "Operations", "Dispatch Officer", "a.nwosu@giglogistics.com", "Lagos"),
        ("Amina Usman", "Human Resources", "HR Manager", "a.usman@giglogistics.com", "Abuja"),
        ("Rotimi Adeyemi", "Finance", "Accountant", "r.adeyemi@giglogistics.com", "Lagos"),
        ("Oluwaseun Balogun", "Technology", "Head of Technology", "o.balogun@giglogistics.com", "Lagos"),
        ("Tolulope Ogunwale", "Customer Service", "Customer Service Manager", "t.ogunwale@giglogistics.com", "Lagos"),
        ("Damilola Akintola", "Sales", "Sales Manager", "d.akintola@giglogistics.com", "Lagos"),
        ("Lara Oduola", "Marketing", "Marketing Manager", "l.oduola@giglogistics.com", "Lagos")
    ]
    
    for emp in sample_employees:
        name, dept, role, email, loc = emp
        employees[name] = {
            'department': dept,
            'role': role,
            'email': email,
            'location': loc
        }
        if dept in department_members:
            department_members[dept].append(name)

def check_employee():
    """Check if the entered name is an employee"""
    name = name_entry.get().strip()
    department = department_entry.get().strip()
    
    if not name or not department:
        messagebox.showwarning("Missing Info", "Please enter both name and department")
        return
    
    # Check if employee exists
    if name in employees:
        emp_data = employees[name]
        
        # Check if department matches
        if emp_data['department'].lower() == department.lower():
            # Show welcome message
            messagebox.showinfo("Welcome!", f"Welcome {name}!\n\n"
                               f"Role: {emp_data['role']}\n"
                               f"Department: {emp_data['department']}\n"
                               f"Location: {emp_data['location']}")
            
            # Show department members
            show_department_members(emp_data['department'])
        else:
            messagebox.showerror("Department Error", 
                               f"{name} works in {emp_data['department']} department,\n"
                               f"not {department} department.")
    else:
        messagebox.showinfo("Not Found", 
                          f"Sorry, {name} is not an employee of GIG Logistics.\n"
                          f"Please check the name and try again.")

def show_department_members(department):
    """Create a new window to show all members of a department"""
    members = department_members.get(department, [])
    
    if not members:
        messagebox.showinfo("No Members", f"No members found in {department} department")
        return
    
    # Create new window
    dept_window = tk.Toplevel(root)
    dept_window.title(f"{department} Department")
    dept_window.geometry("400x400")
    dept_window.configure(bg="#f0f0f0")
    
    # Title
    title = tk.Label(dept_window, text=f"{department} Department", 
                     font=("Arial", 16, "bold"), bg="#f0f0f0", fg="blue")
    title.pack(pady=10)
    
    # Number of employees
    count_label = tk.Label(dept_window, text=f"Total: {len(members)} employees", 
                          font=("Arial", 10), bg="#f0f0f0")
    count_label.pack(pady=5)
    
    # Separator
    separator = tk.Label(dept_window, text="-" * 30, bg="#f0f0f0")
    separator.pack()
    
    # List of employees
    members_text = ""
    for i, member in enumerate(members, 1):
        members_text += f"{i}. {member}\n"
    
    members_label = tk.Label(dept_window, text=members_text, 
                             font=("Arial", 11), bg="#f0f0f0", justify="left")
    members_label.pack(pady=10)
    
    # Close button
    close_btn = tk.Button(dept_window, text="Close", command=dept_window.destroy,
                          bg="red", fg="white", font=("Arial", 10))
    close_btn.pack(pady=10)

def clear_fields():
    """Clear the input fields"""
    name_entry.delete(0, tk.END)
    department_entry.delete(0, tk.END)

# Load employee data
load_employees_from_csv()

# Create main window
root = tk.Tk()
root.title("GIG Logistics - Employee Verification System")
root.geometry("500x500")
root.configure(bg="#f0f0f0")

# Header
header = tk.Label(root, text="GIG LOGISTICS", font=("Arial", 20, "bold"), 
                  bg="#003366", fg="white", pady=10)
header.pack(fill="x")

subheader = tk.Label(root, text="Employee Verification System", 
                     font=("Arial", 12), bg="#f0f0f0", fg="#003366")
subheader.pack(pady=10)

# Instructions
instructions = tk.Label(root, text="Please enter employee details below:", 
                        font=("Arial", 10), bg="#f0f0f0")
instructions.pack(pady=10)

# Name input
name_label = tk.Label(root, text="Employee Name:", font=("Arial", 11), bg="#f0f0f0")
name_label.pack()
name_entry = tk.Entry(root, font=("Arial", 11), width=35)
name_entry.pack(pady=5)

# Department input
dept_label = tk.Label(root, text="Department:", font=("Arial", 11), bg="#f0f0f0")
dept_label.pack()
department_entry = tk.Entry(root, font=("Arial", 11), width=35)
department_entry.pack(pady=5)

# Buttons frame
button_frame = tk.Frame(root, bg="#f0f0f0")
button_frame.pack(pady=20)

# Verify button
verify_btn = tk.Button(button_frame, text="Verify Employee", command=check_employee,
                       bg="green", fg="white", font=("Arial", 11), padx=15, pady=5)
verify_btn.pack(side="left", padx=10)

# Clear button
clear_btn = tk.Button(button_frame, text="Clear", command=clear_fields,
                      bg="orange", fg="white", font=("Arial", 11), padx=15, pady=5)
clear_btn.pack(side="left", padx=10)

# Exit button
exit_btn = tk.Button(button_frame, text="Exit", command=root.quit,
                     bg="red", fg="white", font=("Arial", 11), padx=15, pady=5)
exit_btn.pack(side="left", padx=10)

# Info box
info_text = """
Available Departments:
• Operations
• Finance
• Human Resources
• Technology
• Customer Service
• Sales
• Marketing

Tip: Enter the full name as shown in the list
"""

info_box = tk.Label(root, text=info_text, font=("Arial", 9), 
                    bg="#e6f3ff", fg="#333", justify="left", padx=10, pady=10)
info_box.pack(pady=10)

# Footer
footer = tk.Label(root, text="© 2024 GIG Logistics - All Rights Reserved", 
                  font=("Arial", 8), bg="#f0f0f0", fg="#666")
footer.pack(side="bottom", pady=5)

# Start the program
root.mainloop()
```
