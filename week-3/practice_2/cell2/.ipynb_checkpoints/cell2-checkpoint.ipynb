```python
# cell_2.py - Simple login form

import tkinter as tk
from tkinter import messagebox

def show_welcome():
    # Get username and password
    username = username_entry.get()
    password = password_entry.get()
    
    # Check if correct
    if username == "Mary" and password == "cos102":
        # Create welcome window
        welcome_window = tk.Toplevel(root)
        welcome_window.title("Welcome")
        welcome_window.geometry("400x150")
        
        label1 = tk.Label(welcome_window, text=f"Welcome {username}!", font=("Arial", 14))
        label1.pack(pady=10)
        
        label2 = tk.Label(welcome_window, text="This is Python GUI with tkinter")
        label2.pack(pady=5)
        
        close_btn = tk.Button(welcome_window, text="Close", command=welcome_window.destroy)
        close_btn.pack(pady=10)
    else:
        messagebox.showerror("Login Error", "Invalid username or password")

# Create main window
root = tk.Tk()
root.title("Login Form")
root.geometry("400x250")

# Title
title = tk.Label(root, text="Login System", font=("Arial", 16, "bold"))
title.pack(pady=10)

# Username
username_label = tk.Label(root, text="Username:")
username_label.pack()
username_entry = tk.Entry(root, width=30)
username_entry.pack(pady=5)

# Password
password_label = tk.Label(root, text="Password:")
password_label.pack()
password_entry = tk.Entry(root, show="*", width=30)
password_entry.pack(pady=5)

# Submit button
submit_button = tk.Button(root, text="Submit", command=show_welcome, bg="blue", fg="white")
submit_button.pack(pady=20)

# Start program
root.mainloop()
```
