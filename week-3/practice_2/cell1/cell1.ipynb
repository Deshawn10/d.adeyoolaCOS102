```python
# cell_1.py - Simple window with a button

import tkinter as tk
from tkinter import messagebox

def button_click():
    # Show a message box when button is clicked
    messagebox.showinfo("Info", "Welcome to COS 102 GUI App!\n")
    
    # Ask user a question
    result = messagebox.askyesno("Confirmation", "Do you want to continue?")
    if result:
        messagebox.showinfo("Info", "You said Yes!")
    else:
        messagebox.showinfo("Info", "You said No!")

# Create main window
root = tk.Tk()
root.title("Home Page")
root.geometry("300x150")

# Add a label
label = tk.Label(root, text="Hello Friend\n")
label.pack()

# Add a button
button = tk.Button(root, text="Click Me!", command=button_click)
button.pack()

# Style the button
button.config(fg="red", bg="yellow")

# Start the program
root.mainloop()
```
